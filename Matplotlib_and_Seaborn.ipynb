```python
# Using Multiple Lines plotting
years = range(2014, 2022)
oranges = [0.883, 0.912, 0.921, 0.927, 0.939, 0.941, 0.954, 0.966]
mangos = [0.973, 0.969, 0.951, 0.943, 0.938, 0.928, 0.917, 0.913]
```

```python
plt.plot(years, oranges)
plt.plot(years, mangos)
plt.xlabel('Year')
plt.ylabel('Production (tons per hectare)')
plt.title("Fruit Production in Florida") # adding title
plt.legend(['Oranges', 'Mangos']); # adding legend
```

```python
# adding the line markers
plt.plot(years, oranges, marker='v')
plt.plot(years, mangos, marker='p')
plt.xlabel('Year')
plt.ylabel('Production (tons per hectare)')
plt.title("Fruit Production in Florida")
plt.legend(['Oranges', 'Mangos']); 
```

```python
plt.plot(years, oranges, 'or')
plt.title("Production of Oranges (tons per hectare)");
```

```python
# Using Seaborn to change style
sns.set_style("darkgrid") # this style is global
```

```python
# The `fmt` argument provides a shorthand for specifying the marker shape, line style, and line color. 
# It can be provided as the third argument to `plt.plot`.
# fmt = '[marker][line][color]'
```

```python
plt.plot(years, oranges, 's-b')
plt.plot(years, mangos, 'o--r')
plt.xlabel('Year')
plt.ylabel('Production (tons per hectare)')
plt.title("Fruit Production in Florida")
plt.legend(['Oranges', 'Mangos']); 
```

```python
# Scatter Plots on a two dimentional grid.
# Using "iris" dataset (The Iris dataset is included with the Seaborn library and can be loaded as a Pandas data frame).
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# visualizing the relationship between sepal length and sepal width. Line chart not a good approach
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
# using scatter plots with hue argument(seaborn module)
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species, 
                s=150);
```

```python
# Adding title and formattiong chart size
plt.figure(figsize=(10, 5))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species,
                s=100);
```

```python
# Histogram -> Using the plt.hist function to create a histogram.
# Load data into a Pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.describe()
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
```

```python
# Drawing multiple histograms in a single chart
import numpy as np
```

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
```

```python
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```
